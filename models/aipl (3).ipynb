```python
import numpy as np
import pandas as pd
```

```python
d1=pd.read_csv(r"C:\Users\Saraf Nihal chandra\Desktop\last project 1\models\101.csv")
d2=pd.read_csv(r"C:\Users\Saraf Nihal chandra\Desktop\last project 1\models\111.csv.csv")
d1
d1.to_csv("101.csv")
```

```python
replacements={
    "Deccan Chargers":"Sunrisers Hyderabad",
    "Kings XI Punjab":"Punjab Kings",
    "Delhi Daredevils":"Delhi Capitals",
    "Royal Challengers Bangalore":"Royal Challengers Bengaluru"
}
```

```python
d2.replace(replacements,inplace=True)
d2
d2.to_csv("111.csv")
```

```python
total_runs=d2.groupby(['id','innings']).sum()['total_run'].add(1).reset_index()
total_runs
```

```python
total_runs = total_runs[total_runs['innings']==1]
```

```python
match =d1.merge(total_runs[['id','total_run']],left_on='id',right_on='id')
match
```

```python
teams=['Royal Challengers Bengaluru',
       'Chennai Super Kings',
       'Delhi Capitals',
       'Mumbai Indians',
       'Sunrisers Hyderabad',
       'Kolkata Knight Riders',
       'Punjab Kings',
       'Rajasthan Royals',
       'Lucknow Super Giants',
       'Gujarat Titans'
      ]
```

```python
match=match[match['team1'].isin(teams)]
match=match[match['team2'].isin(teams)]
```

```python
match['team1'].value_counts()
match['team2'].value_counts()
```

```python
match=match[['id','city','winner','total_run','team1','team2']]
```

```python
second_innings=match.merge(d2,on='id')
second_innings
```

```python
second_innings=second_innings[second_innings['innings']==2]
second_innings
```

```python
second_innings['current_score'] = second_innings.groupby('id')['total_run_y'].cumsum()
second_innings
```

```python
second_innings['runs_left'] = second_innings['total_run_x'] - second_innings['current_score']
second_innings
```

```python
second_innings['balls_left'] = 126 - (second_innings['over']*6 +second_innings['ball'])
second_innings
```

```python

columns_to_check = ['current_score', 'runs_left', 'balls_left']

# Check how many values in each column are negative
negative_counts = second_innings[columns_to_check].lt(0).sum()

# Display the counts
print("Number of negative values in each column:")
print(negative_counts)
```

```python
second_innings['player_dismissed'] = second_innings['player_dismissed'].fillna(0)
#second_innings['player_dismissed'] = second_innings['player_dismissed'].apply(lambda x:"1" if x!="0" else "0")
#second_innings['player_dismissed'] = second_innings['player_dismissed'].astype('int')
wickets = second_innings.groupby('id')['is_wicket'].cumsum()
second_innings['wickets_remaining'] = 10-wickets
second_innings
```

```python
second_innings
```

```python
second_innings['crr'] = round((second_innings['current_score']*6)/(120 - second_innings['balls_left']),2)
second_innings['rrr'] = round((second_innings['runs_left']*6)/(second_innings['balls_left']),2)
second_innings
```

```python
def winner(team):
    return 1 if team['batting_team'] == team['winner'] else 0
```

```python
second_innings['results'] = second_innings.apply(winner,axis=1)
second_innings
```

```python
required_data = second_innings[['batting_team','bowling_team','city','runs_left','balls_left','wickets_remaining','total_run_x','crr','rrr','results']]
required_data
```

```python
required_data = required_data.replace([np.inf, -np.inf], np.nan)
required_data = required_data.dropna()
required_data = required_data[required_data['crr'] >= 0]
```

```python
columns_to_check = ['runs_left', 'balls_left', 'wickets_remaining', 'total_run_x',  'results']

required_data = required_data[~(required_data[columns_to_check] < 0).any(axis=1)]
```

```python
required_data
```

```python
numeric = required_data[['runs_left', 'balls_left', 'wickets_remaining', 'total_run_x', 'crr', 'rrr', 'results']]
has_infinite_values = np.any(np.isinf(numeric))

# Check for negative values in the entire DataFrame
has_negative_values = (numeric < 0).any().any()

# Check for null values in the entire DataFrame
has_null_values = numeric.isnull().any().any()

# Display the results
print(f"Has Infinite Values: {has_infinite_values}")
print(f"Has Negative Values: {has_negative_values}")
print(f"Has Null Values: {has_null_values}")
```

```python
required_data.to_csv("cd.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = required_data.drop(['crr','rrr','results'],axis=1)
y = required_data['results']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=100)
y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder 
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],
remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(x_test)[0]
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```

```python
from sklearn.ensemble import RandomForestClassifier

ra_pipe = Pipeline([
    ('step1',trf),
    ('step2',RandomForestClassifier())
])

ra_pipe.fit(x_train,y_train)

ra_y_pred = ra_pipe.predict(x_test)


accuracy_score(y_test,ra_y_pred)
```

```python
ra_pipe.predict_proba(x_test)[0]
```
